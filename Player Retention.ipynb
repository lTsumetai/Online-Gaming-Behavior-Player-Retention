### Section 4. Data Analysis and Exploratory Data Analysis (EDA) for SMART framework and the main problem of case study

##### What factors influence player engagement behavior?
To answer this question, we analyze whether there is a significant difference in player activity between different engagement groups.

in particular, we want to understand :
> Do highly engaged players spend more time playing compared to low-engagement players?


We can gather that there is a variable called 
> EngagementLevel: Indicates the level of player engagement categorized as 'High', 'Medium', or 'Low'.

we can identify what it means and how we will use this variable to answer our factors  

```python
# Group by EngagementLevel and calculate average sessions
engagement_summary = (
    Online_gaming_df.groupby('EngagementLevel')['SessionsPerWeek']
      .mean()
      .reset_index()
      .rename(columns={'SessionsPerWeek': 'AvgSessionsPerWeek'})
      .sort_values(by='AvgSessionsPerWeek')
)

engagement_summary
```

```python
# Bar chart: visualiazing average sessions per week by engagement level
plt.figure(figsize=(10, 6))
plt.bar(engagement_summary['EngagementLevel'], engagement_summary['AvgSessionsPerWeek'], color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Average Sessions Per Week by Engagement Level')

plt.tight_layout()
plt.show()
```

The results show that the higher engagement level players have, the more sessions they have per week.

From the results we can confirm that EngagementLevel Variable is a summary label that's represents player activity.

since EngagementLevel Confirms a summary label, we can use it to perform Statistical Significance with the current hypothesis :

>$H_0$: There is no difference in mean playtime between low and high engagement players.

>$H_1$: There is a significant difference in mean playtime between low and high engagement players.

We will use confidence level of 95%. which means p-value = 0.05

```python
# grouping/filtering data using labels based on EngagementLevel
low  = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'Low']['PlayTimeHours']
high = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'High']['PlayTimeHours']
```

```python
print("Low Engagement PlayTime:")
print(low)
print("\nHigh Engagement PlayTime:")
print(high)
```

```python
# Using Two-Sample T-test
t_stat, p_value = stats.ttest_ind(low, high)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

print("Mean Playtime (Low):", low.mean())
print("Mean Playtime (High):", high.mean())
print()

if p_value < alpha:
    print("Reject H0 : There is a significant difference in playtime.")
else:
    print("Fail to reject H0 : No significant difference detected.")
```

```python
# Bar chart: Mean PlayTimeHours by Engagement Level
means = [low.mean(), high.mean()]
labels = ['Low Engagement', 'High Engagement']
colors = ['#f28b82', '#81c995']

fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(labels, means, color=colors, edgecolor='white', width=0.5)
ax.bar_label(bars, fmt='%.2f', padding=5, fontsize=11)
ax.set_title('Mean PlayTimeHours: Low vs High Engagement', fontsize=13, fontweight='bold')
ax.set_ylabel('Mean Play Time (Hours)')
ax.set_ylim(0, max(means) * 1.15)

plt.tight_layout()
plt.show()
```

Based on Two Sample T-test. the result shows that the average playtime for low engagement players (12.10 hours) and high engagement players (12.07 hours) is nearly identical. meaning that there is no significant difference in playtime between low and high engagement players.


##### What do players play?


To understand *what* players play, we examine the distribution of game genres across engagement levels.
This tells us which genres are associated with higher or lower engagement — directly informing what game experiences to invest in.

Hence, we want to know :
> Do certain game genres attract more highly engaged players?

therefore, we can analyze if there is a significant correlation between GameGenre and EngagementLevel. for this problem, we will use a **Chi-Square Test of Independence** since both variables are categorical.

we can use to perforce statistical significance with the current hyphotehsis

> $H_0$: `GameGenre` and `EngagementLevel` are independent (no association).  
> $H_1$: `GameGenre` and `EngagementLevel` are associated.

and as usual, the confidence level will be: 95% (α = 0.05)

```python
# Cross-tabulation: GameGenre vs EngagementLevel
genre_engagement = pd.crosstab(Online_gaming_df['GameGenre'], Online_gaming_df['EngagementLevel'])
print('Contingency Table (GameGenre vs EngagementLevel):')
print(genre_engagement)
```

```python
# Chi-Square Test of Independence
chi2, p_chi, dof, expected = stats.chi2_contingency(genre_engagement)

print(f'Chi-Square Statistic : {chi2:.4f}')
print(f'P-value              : {p_chi:.4f}')
print(f'Degrees of Freedom   : {dof}')
print()
if p_chi < 0.05:
    print('Reject H0: GameGenre and EngagementLevel are significantly associated.')
else:
    print('Fail to reject H0: No significant association between GameGenre and EngagementLevel.')
```

```python
# Visualize: Genre distribution across Engagement Levels (stacked bar)
genre_pct = genre_engagement.div(genre_engagement.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Count plot
genre_engagement.plot(kind='bar', ax=axes[0], colormap='Set2', edgecolor='white')
axes[0].set_title('GameGenre vs EngagementLevel (Count)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Game Genre')
axes[0].set_ylabel('Player Count')
axes[0].tick_params(axis='x', rotation=30)
axes[0].legend(title='Engagement Level')

# Percentage plot
genre_pct.plot(kind='bar', stacked=True, ax=axes[1], colormap='Set2', edgecolor='white')
axes[1].set_title('GameGenre vs EngagementLevel (% Share)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Game Genre')
axes[1].set_ylabel('Percentage (%)')
axes[1].tick_params(axis='x', rotation=30)
axes[1].legend(title='Engagement Level', loc='upper right')

plt.tight_layout()
plt.show()
```
