### Section 5. Data Analysis and Exploratory Data Analysis (EDA) for SMART framework and the main problem of case study

##### What factors influence player engagement behavior?
To answer this question, we analyze whether there is a significant difference in player activity between different engagement groups.

in particular, we want to understand :
> Do highly engaged players spend more time playing compared to low-engagement players?


We can gather that there is a variable called 
> EngagementLevel: Indicates the level of player engagement categorized as 'High', 'Medium', or 'Low'.

we can identify what it means and how we will use this variable to answer our factors  

```python
# Group by EngagementLevel and calculate average sessions
engagement_summary = (
    Online_gaming_df.groupby('EngagementLevel')['SessionsPerWeek']
      .mean()
      .reset_index()
      .rename(columns={'SessionsPerWeek': 'AvgSessionsPerWeek'})
      .sort_values(by='AvgSessionsPerWeek')
)

engagement_summary
```

The results show that the higher engagement level players have, the more sessions they have per week.

From the results we can confirm that EngagementLevel Variable is a summary label that's represents player activity.

since EngagementLevel Confirms a summary label, we can use it to perform Statistical Significance with the current hypothesis :

>$H_0$: There is no difference in mean playtime between low and high engagement players.

>$H_1$: There is a significant difference in mean playtime between low and high engagement players.

We will use confidence level of 95%. which means p-value = 0.05

```python
# grouping/filtering data using labels based on EngagementLevel
low  = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'Low']['PlayTimeHours']
high = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'High']['PlayTimeHours']
```

```python
print("Low Engagement PlayTime:")
print(low)
print("\nHigh Engagement PlayTime:")
print(high)
```

```python
# Using Two-Sample T-test
t_stat, p_value = stats.ttest_ind(low, high)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

print("Mean Playtime (Low):", low.mean())
print("Mean Playtime (High):", high.mean())
print()

if p_value < alpha:
    print("Reject H0 : There is a significant difference in playtime.")
else:
    print("Fail to reject H0 : No significant difference detected.")
```

Based on Two Sample T-test. the result shows that the average playtime for low engagement players (12.10 hours) and high engagement players (12.07 hours) is nearly identical. meaning that there is no significant difference in playtime between low and high engagement players.

We can infer from the fact that gamers with low and high levels of engagement spend almost the same amount of time playing. It indicates that a stable player count is already being achieved.

##### Which engagement group has the highest average session frequency? 
To answer this question, we can gather Group players by EngagementLevel and calculate the average sessions per week. This confirms whether the engagement label reflects actual play frequency

```python
result = Online_gaming_df.groupby('EngagementLevel').agg({'SessionsPerWeek': 'mean'})
result = result.reindex(['Low', 'Medium', 'High'])
result
```

```python
result.plot(kind='bar', legend=False, color=['#3266ad', '#5b9ee3', '#1d9e75'], rot=0)
plt.title('Avg Sessions per Week by Engagement Level')
plt.ylabel('Avg Sessions per Week')
plt.xlabel('Engagement Level')
plt.show()
```

##### Which Genre has the most player count? 
To answer this question, we can show the proportion of players in each game genre. This helps understand which genre dominates the playerbase.

```python
result = Online_gaming_df.groupby('GameGenre').agg({'PlayerID': 'count'})
result = result.rename(columns={'PlayerID': 'PlayerCount'})
result
```

```python
result.plot(kind='pie', y='PlayerCount', autopct='%.1f%%', legend=False, ylabel='')
plt.title('Player Count by Game Genre')
plt.show()
```

##### How was the rate of the in-game purchases for each genre ?
To  answer this question, we can Calculate the percentage of players who made in-game purchases for each genre. This gives a monetization overview across game types.

```python
result = Online_gaming_df.groupby('GameGenre').agg({'InGamePurchases': 'mean'}) * 100
result = result.rename(columns={'InGamePurchases': 'PurchaseRate_%'})
result = result.sort_values('PurchaseRate_%', ascending=False)
result
```

```python
result.plot(kind='barh', legend=False, color='#5b9ee3')
plt.title('In-Game Purchase Rate by Genre')
plt.xlabel('Purchase Rate (%)')
plt.show()
```

##### What are the playtime distributions between Low vs High Engagement ?
To answer this question, we can Compare the playtime distribution between Low and High engagement players. This supports the T-test hypothesis that engagement level does not affect hours played.

```python
low  = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'Low']['PlayTimeHours']
high = Online_gaming_df[Online_gaming_df['EngagementLevel'] == 'High']['PlayTimeHours']

t_stat, p_value = stats.ttest_ind(low, high)
print('Mean Low :', low.mean().round(2))
print('Mean High:', high.mean().round(2))
print('P-value  :', round(p_value, 4))
```

```python
plt.hist(low,  bins=30, alpha=0.6, color='#e24b4a', label='Low')
plt.hist(high, bins=30, alpha=0.6, color='#1d9e75', label='High')
plt.title('PlayTime Distribution: Low vs High Engagement')
plt.xlabel('Play Time (hours)')
plt.ylabel('Number of Players')
plt.legend()
plt.show()
```

##### How many sessions per week do players typically play?
To answer this, we calculate the mean, median, and mode of SessionsPerWeek. This helps the platform decide how often to send notifications or release weekly challenges.

```python
result = Online_gaming_df['SessionsPerWeek']

print('Mean  :', result.mean().round(2))
print('Median:', result.median())
print('Mode  :', result.mode()[0])
```

```python
result.plot(kind='hist', bins=30, color='#5b9ee3', edgecolor='white')
plt.axvline(result.mean(),   color='red',    linestyle='--', label=f'Mean {result.mean():.1f}')
plt.axvline(result.median(), color='orange', linestyle='--', label=f'Median {result.median():.1f}')
plt.title('Distribution of Sessions per Week')
plt.xlabel('Sessions per Week')
plt.legend()
plt.show()
```
